### Query para criar o dataset único

```python
dataset_unico = pd.read_sql(
    """
        SELECT V.*, CC.NOME_CLIENTE, CP.MATERIAL, CEV.ESCRITORIO_DE_VENDAS, CEqV.EQUIPE_DE_VENDAS, CV.NOME_REPRESENTANTE_PARCEIRO, V.QTDE_VENDIDA * V.VALOR_UNITARIO AS VALOR_VENDIDO 
        FROM
            (SELECT * FROM Vendas_2017
            UNION ALL
            SELECT * FROM Vendas_2018
            UNION ALL
            SELECT * FROM Vendas_2019
            UNION ALL
            SELECT * FROM Vendas_2020
            UNION ALL
            SELECT * FROM Vendas_2021
            UNION ALL
            SELECT * FROM Vendas_2022) V
            
        INNER JOIN
            (SELECT CC.*, row_number() OVER (PARTITION BY COD_CLIENTE) as REGISTROS_CC FROM CADASTRO_CLIENTES CC) CC
        ON V.COD_CLIENTE = CC.COD_CLIENTE and REGISTROS_CC = 1
        
        INNER JOIN
            (SELECT CP.*, row_number() OVER (PARTITION BY CODIGO_MATERIAL) as REGISTROS_CP FROM CADASTRO_PRODUTOS CP) CP
        ON V.CODIGO_MATERIAL = CP.CODIGO_MATERIAL and REGISTROS_CP = 1
        
        INNER JOIN
            (SELECT CEV.*, row_number() OVER (PARTITION BY COD_ESCRITORIO_DE_VENDAS) as REGISTROS_CEV FROM CADASTRO_ESCRITORIO_VENDAS CEV) CEV
        ON V.COD_ESCRITORIO_DE_VENDAS = CEV.COD_ESCRITORIO_DE_VENDAS and REGISTROS_CEV = 1
        
        INNER JOIN
            (SELECT CEqV.*, row_number() OVER (PARTITION BY COD_EQUIPE_DE_VENDAS) as REGISTROS_CEqV FROM CADASTRO_EQUIPE_VENDAS CEqV) CEqV
        ON V.COD_EQUIPE_DE_VENDAS = CEqV.COD_EQUIPE_DE_VENDAS and REGISTROS_CEqV = 1
        
        LEFT JOIN
            (SELECT CV.*, row_number() OVER (PARTITION BY COD_REPRESENTANTE_PARCEIRO) as REGISTROS_CV FROM CADASTRO_VENDEDOR CV) CV
        ON V.COD_REPRESENTANTE_PARCEIRO = CV.COD_REPRESENTANTE_PARCEIRO and REGISTROS_CV = 1
        
        ORDER BY V."DATA DA VENDA" ASC
    """,
    conn
)
```

# Estimação do volume de vendas para o ano de 2023

## Treinando o modelo com rede neural RNN (Recurrent Neural Network - Rede Neural Recorrente) com arquitetura LSTM (Long Short Term Memory) para dados sequenciais
## - Método 1: GROUP BY para ter o volume de vendas total por mês de cada ano de toda a base e depois um gráfico para visualizar (somente treino com todos os dados disponíveis, sem test set)
## - Método 2: Retiro os últimos 6 meses da base para o test set para ver o desempenho do modelo

```python
dataset_train_1 = pd.read_sql(
    """
        SELECT strftime("%Y-%m", "DATA DA VENDA") AS DATA_DA_VENDA, SUM(QTDE_VENDIDA) AS QTDE_VENDIDA FROM dataset_unico
        GROUP BY DATA_DA_VENDA
    """,
    conn
)

dataset_train_2 = pd.read_sql(
    """
        SELECT strftime("%Y-%m", "DATA DA VENDA") AS DATA_DA_VENDA, SUM(QTDE_VENDIDA) AS QTDE_VENDIDA FROM dataset_unico
        WHERE DATA_DA_VENDA < "2022-01"
        GROUP BY DATA_DA_VENDA
    """,
    conn
)
```

```python
training_set_1 = dataset_train_1.iloc[:,1:2].values
training_set_2 = dataset_train_2.iloc[:,1:2].values
```

```python
plt.figure(figsize=(16,8))
plt.plot(training_set_2, color='#96C139', label='Volume real de vendas em quantidade absoluta')
plt.title('Volume real de vendas')
plt.xlabel('Tempo (Meses)')
plt.ylabel('Volume de vendas (em milhões)')
plt.legend()
plt.show()
```

### Feature scaling com o Scikit-Learn

```python
from sklearn.preprocessing import MinMaxScaler

sc = MinMaxScaler(feature_range = (0,1))
training_set_scaled_2 = sc.fit_transform(training_set_2)
```

### Criando uma estrutura de dados de forma sequencial com janelas de 12 meses e com 1 output em cada janela (prevendo o mês seguinte)

```python
janela = 12

x_train_2 = []
y_train_2 = []
for i in range(janela, training_set_2.shape[0]):
    x_train_2.append(training_set_scaled_2[i-janela:i,0])
    y_train_2.append(training_set_scaled_2[i,0])
x_train_2, y_train_2 = np.array(x_train_2), np.array(y_train_2)

# Fazendo o reshape
x_train_2 = np.reshape(x_train_2, (x_train_2.shape[0], x_train_2.shape[1], 1))

# data_path = "/Estimação - LSTM"
data_path = "C:\\Users\\rd-co\\Desktop\\Case Ourofino\\Cientista de Dados\\Estimação - LSTM"

# Construindo a estrutura RNN (Recurrent Neural Network)
# Importando as libraries e packages do Keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
from keras.callbacks import ModelCheckpoint, EarlyStopping

# Inicializando o RNN
regressor = Sequential()

# Adicionando os layers LSTM e regularização com a técnica Dropout
regressor.add(LSTM(units = 500, return_sequences = True, input_shape = (x_train_2.shape[1], 1)))
regressor.add(Dropout(0.2))
regressor.add(LSTM(units = 300, return_sequences = False))
regressor.add(Dropout(0.2))

# Adicionando o output layer
regressor.add(Dense(units = 1, activation='linear'))

# Fazendo o compile do RNN
regressor.compile(optimizer = "adam", loss = "mean_squared_error", metrics=['mean_absolute_error'])       # Regressão, pois a variável em questão é quantitativa

# Checkpoint e earlyStopping para salvar os melhores regressores com as menores perdas (loss)
checkpointer = ModelCheckpoint(filepath=data_path + "\\regressor-{epoch:02d}-{loss:.4f}.hdf5",
                               monitor="loss", mode="min", save_best_only=True, verbose=1)

earlyStop = EarlyStopping(
        monitor='loss',
        patience=60,
        verbose=1,
        mode='min',
        restore_best_weights=True
)

# Fazendo o fit do RNN no training set e salvando em disco os melhores regressores
regressor.fit(x_train_2, y_train_2, epochs = 300, batch_size = 32, callbacks=[earlyStop])
```

```python
real_values_2 = dataset_test_2.iloc[:,1:2].values

dataset_total_2 = pd.concat((dataset_train_2["QTDE_VENDIDA"], dataset_test_2["QTDE_VENDIDA"]), axis = 0)
inputs_2 = dataset_total_2[len(dataset_total_2) - len(dataset_test_2) - janela:].values
inputs_2 = inputs_2.reshape(-1,1)
inputs_2 = sc.transform(inputs_2)

x_test_2 = []
for i in range(janela, janela + dataset_test_2.shape[0]):
    x_test_2.append(inputs_2[i-janela:i,0])
x_test_2 = np.array(x_test_2)
x_test_2 = np.reshape(x_test_2, (x_test_2.shape[0], x_test_2.shape[1], 1))

predicted_value_2 = regressor.predict(x_test_2)
predicted_value_2 = sc.inverse_transform(predicted_value_2)
predicted_value_2 = np.round(predicted_value_2)
```

### Estatística F
### Análise de variância de um fator:
### As variáveis que mais contribuem para a formação de pelo menos um dos clusters

```python
def teste_f_kmeans(kmeans, dataframe):
    
    variaveis = dataframe.columns

    centroides = pd.DataFrame(kmeans.cluster_centers_)
    centroides.columns = dataframe.columns
    centroides
    
    print("Centróides: \n", centroides ,"\n")

    df = dataframe[variaveis]

    unique, counts = np.unique(kmeans.labels_, return_counts=True)

    dic = dict(zip(unique, counts))

    qnt_clusters = kmeans.n_clusters

    observacoes = len(kmeans.labels_)

    df['cluster'] = kmeans.labels_

    output = []

    for variavel in variaveis:

        dic_var={'variavel':variavel}

        # variabilidade entre os grupos

        variabilidade_entre_grupos = np.sum([dic[index]*np.square(observacao - df[variavel].mean()) for index, observacao in enumerate(centroides[variavel])])/(qnt_clusters - 1)

        dic_var['variabilidade_entre_grupos'] = variabilidade_entre_grupos

        variabilidade_dentro_dos_grupos = 0

        for grupo in unique:

            grupo = df[df.cluster == grupo]

            variabilidade_dentro_dos_grupos += np.sum([np.square(observacao - grupo[variavel].mean()) for observacao in grupo[variavel]])/(observacoes - qnt_clusters)

        dic_var['variabilidade_dentro_dos_grupos'] = variabilidade_dentro_dos_grupos

        dic_var['F'] =  dic_var['variabilidade_entre_grupos']/dic_var['variabilidade_dentro_dos_grupos']
        
        dic_var['sig F'] =  1 - stats.f.cdf(dic_var['F'], qnt_clusters - 1, observacoes - qnt_clusters)

        output.append(dic_var)

    df = pd.DataFrame(output)
    
    print(df)

    return df
```

### Os valores da estatística F são bastante sensíveis ao tamanho da amostra

```python
output = teste_f_kmeans(kmeans, clientes_valor_cluster)
```
